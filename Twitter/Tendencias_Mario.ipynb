```python
import re

def obtenerTendencias(df_tweets, dia = None):
    if dia != None:
        filtro = [True if re.match(dia, fecha) != None else False for fecha in df_tweets["created_at"]]
        df_tweets = df_tweets[filtro]
    tendencias = {}

    for index, row in df_tweets.iterrows():
        for word_start in row['text'].split():
            word = word_start.lower()
            if word not in ignore:
                try:
                    tendencias[word] += 1 
                except:
                    tendencias[word] = 1
    return sorted(tendencias.items(), key=lambda tup: tup[1], reverse=True)
```
